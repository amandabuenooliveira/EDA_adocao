# Análise de produtos da EDB

```python
from collections import defaultdict
```

Análise de Cesta de Produtos:

```python

def analisar_cesta_produtos_colecao_df_brasil(df_brasil):
    """
    Identifica pares de coleções frequentemente adotadas juntas na mesma escola e ano, usando apenas df_brasil.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.

    Returns:
        dict: Dicionário com pares de coleções e suas contagens.
    """

    # Agrupa por escola e ano no df_brasil
    grupos = df_brasil.groupby(['CD_ESCOLA', 'NU_ANO_ADOCAO'])

    pares_colecoes = defaultdict(int)

    for _, grupo in grupos:
        # Converte todos os valores para string, tratando NaN
        colecoes = grupo['NM_COLECAO'].astype(str).fillna('').unique()

        if len(colecoes) > 1:
            for i in range(len(colecoes)):
                for j in range(i + 1, len(colecoes)):
                    par = tuple(sorted((colecoes[i], colecoes[j]))) # Ordena para evitar duplicações
                    pares_colecoes[par] += 1

    return pares_colecoes

# Supondo que seu DataFrame se chama df_brasil
pares_colecoes_adotadas_brasil = analisar_cesta_produtos_colecao_df_brasil(df_brasil)

# Exibe os pares mais frequentes
pares_ordenados_colecoes_brasil = sorted(pares_colecoes_adotadas_brasil.items(), key=lambda x: x[1], reverse=True)
print("Pares de coleções mais adotadas juntas (df_brasil):")
for par, contagem in pares_ordenados_colecoes_brasil[:20]: # Exibe os 10 pares mais frequentes
    print(f"{par}: {contagem} vezes")
```

Análise de Ciclo de Vida do Produto

```python
def analisar_ciclo_vida_top_10_colecoes_2024(df_brasil):
    """
    Analisa o ciclo de vida das 10 coleções mais adotadas em 2024.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.
    """

    # 1. Filtrar dados para 2024
    df_2024 = df_brasil[df_brasil['NU_ANO_ADOCAO'] == 2024]

    # 2. Identificar as 10 coleções mais adotadas em 2024
    top_10_colecoes_2024 = df_2024['NM_COLECAO'].value_counts().nlargest(10).index.tolist()

    # 3. Analisar o ciclo de vida dessas coleções
    for colecao in top_10_colecoes_2024:
        dados_colecao = df_brasil[df_brasil['NM_COLECAO'] == colecao].groupby('NU_ANO_ADOCAO')['QT_ADOCAO'].sum().reset_index()

        plt.figure(figsize=(10, 6))
        sns.lineplot(x='NU_ANO_ADOCAO', y='QT_ADOCAO', data=dados_colecao)
        plt.title(f'Adoções da Coleção "{colecao}" ao Longo do Tempo')
        plt.xlabel('Ano de Adoção')
        plt.ylabel('Quantidade de Adoções')
        plt.xticks(dados_colecao['NU_ANO_ADOCAO'].unique()) # Garante que todos os anos sejam exibidos no eixo x
        plt.show()

# Supondo que seu DataFrame se chama df_brasil
analisar_ciclo_vida_top_10_colecoes_2024(df_brasil)
```

Análise adoção da EDB

```python
def analisar_fidelidade_escolas(df_brasil):
    """
    Identifica as escolas que adotam consistentemente os produtos da editora ao longo do tempo.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.

    Returns:
        dict: Dicionário com escolas e a contagem de anos em que adotaram produtos.
    """

    fidelidade_escolas = defaultdict(int)
    escolas_anos = df_brasil.groupby('CD_ESCOLA')['NU_ANO_ADOCAO'].unique()

    for escola, anos in escolas_anos.items():
        fidelidade_escolas[escola] = len(anos)

    return fidelidade_escolas

# Supondo que seu DataFrame se chama df_brasil
fidelidade_escolas = analisar_fidelidade_escolas(df_brasil)

# Exibe as escolas mais fiéis
fidelidade_ordenada = sorted(fidelidade_escolas.items(), key=lambda x: x[1], reverse=True)
print("Escolas mais fiéis (contagem de anos de adoção):")
for escola, anos in fidelidade_ordenada[:10]: # Exibe as 10 escolas mais fiéis
    print(f"Escola {escola}: {anos} anos")
```

Análise de Crescimento das Escolas

```python
def analisar_crescimento_escolas(df_brasil):
    """
    Identifica as escolas com maior crescimento de adoção de um ano para o outro.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.

    Returns:
        pd.DataFrame: DataFrame com escolas, anos e crescimento de adoção.
    """

    dados_escolas = df_brasil.groupby(['CD_ESCOLA', 'NU_ANO_ADOCAO'])['QT_ADOCAO'].sum().reset_index()
    dados_escolas['CRESCIMENTO'] = dados_escolas.groupby('CD_ESCOLA')['QT_ADOCAO'].diff().fillna(0)

    return dados_escolas

# Supondo que seu DataFrame se chama df_brasil
crescimento_escolas = analisar_crescimento_escolas(df_brasil)

# Exibe as escolas com maior crescimento
crescimento_ordenado = crescimento_escolas.sort_values('CRESCIMENTO', ascending=False)
print("Escolas com maior crescimento de adoção:")
print(crescimento_ordenado.head(10)) # Exibe as 10 escolas com maior crescimento
```

Análise de Cancelamento de Adoções

```python
def analisar_cancelamento_adocoes(df_brasil):
    """
    Identifica as escolas que cancelaram adoções de um ano para o outro.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.

    Returns:
        pd.DataFrame: DataFrame com escolas, anos e cancelamento de adoção.
    """

    dados_escolas = df_brasil.groupby(['CD_ESCOLA', 'NU_ANO_ADOCAO'])['QT_ADOCAO'].sum().reset_index()
    dados_escolas['CANCELAMENTO'] = dados_escolas.groupby('CD_ESCOLA')['QT_ADOCAO'].diff().fillna(0)

    cancelamento_escolas = dados_escolas[dados_escolas['CANCELAMENTO'] < 0]

    return cancelamento_escolas

# Supondo que seu DataFrame se chama df_brasil
cancelamento_escolas = analisar_cancelamento_adocoes(df_brasil)

# Exibe as escolas que cancelaram adoções
print("Escolas que cancelaram adoções:")
print(cancelamento_escolas)
```

# Análise GEO adoção EDB 

Análise de Clusterização Geográfica - uso de K-Means para identificar grupos de estados com padrões de adoção semelhantes. 

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

def clusterizar_estados_adocoes(df_brasil):
    """
    Identifica grupos de estados com padrões de adoção semelhantes usando clusterização K-Means.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.
    """

    dados_estados = df_brasil.groupby('NM_ESTADO')['QT_ADOCAO'].sum().reset_index()

    # Normalizar os dados
    scaler = StandardScaler()
    dados_normalizados = scaler.fit_transform(dados_estados[['QT_ADOCAO']])

    # Aplicar K-Means
    kmeans = KMeans(n_clusters=3, random_state=42) # Ajuste o número de clusters conforme necessário
    dados_estados['CLUSTER'] = kmeans.fit_predict(dados_normalizados)

    print(dados_estados)

# Supondo que seu DataFrame se chama df_brasil
clusterizar_estados_adocoes(df_brasil)
```

```python
def comparar_adocoes_regioes(df_brasil):
    """
    Compara as taxas de adoção entre as diferentes regiões geográficas.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.
    """

    dados_regioes = df_brasil.groupby('NM_REGIAO_GEOGRAFICA')['QT_ADOCAO'].sum().reset_index()

    plt.figure(figsize=(10, 6))
    sns.barplot(x='NM_REGIAO_GEOGRAFICA', y='QT_ADOCAO', data=dados_regioes)
    plt.title('Comparação de Adoções por Região Geográfica')
    plt.xlabel('Região Geográfica')
    plt.ylabel('Quantidade de Adoções')
    plt.show()

# Supondo que seu DataFrame se chama df_brasil
comparar_adocoes_regioes(df_brasil)
```

# Analise temporal

Tendências de Adoção ao Longo do Tempo (Segmentado por Matéria)

```python
def analisar_tendencias_adocoes_materia(df_brasil):
    """
    Cria gráficos de linha para visualizar a evolução das adoções ao longo dos anos, segmentando por matéria.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.
    """

    materias = df_brasil['NM_MATERIA'].unique()

    for materia in materias:
        dados_materia = df_brasil[df_brasil['NM_MATERIA'] == materia].groupby('NU_ANO_ADOCAO')['QT_ADOCAO'].sum().reset_index()

        plt.figure(figsize=(10, 6))
        sns.lineplot(x='NU_ANO_ADOCAO', y='QT_ADOCAO', data=dados_materia)
        plt.title(f'Adoções de "{materia}" ao Longo do Tempo')
        plt.xlabel('Ano de Adoção')
        plt.ylabel('Quantidade de Adoções')
        plt.xticks(dados_materia['NU_ANO_ADOCAO'].unique()) # Garante que todos os anos sejam exibidos no eixo x
        plt.show()

# Supondo que seu DataFrame se chama df_brasil
analisar_tendencias_adocoes_materia(df_brasil)
```

```python
def analisar_tendencias_adocoes_segmento(df_brasil):
    """
    Cria gráficos de linha para visualizar a evolução das adoções ao longo dos anos, segmentando por segmento.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.
    """

    segmentos = df_brasil['NM_SEGMENTO'].unique()

    for segmento in segmentos:
        dados_segmento = df_brasil[df_brasil['NM_SEGMENTO'] == segmento].groupby('NU_ANO_ADOCAO')['QT_ADOCAO'].sum().reset_index()

        plt.figure(figsize=(10, 6))
        sns.lineplot(x='NU_ANO_ADOCAO', y='QT_ADOCAO', data=dados_segmento)
        plt.title(f'Adoções de "{segmento}" ao Longo do Tempo')
        plt.xlabel('Ano de Adoção')
        plt.ylabel('Quantidade de Adoções')
        plt.xticks(dados_segmento['NU_ANO_ADOCAO'].unique()) # Garante que todos os anos sejam exibidos no eixo x
        plt.show()

# Supondo que seu DataFrame se chama df_brasil
analisar_tendencias_adocoes_segmento(df_brasil)
```

Uso de Médias Móveis para Suavizar as Flutuações - A média móvel suaviza as flutuações nos dados, facilitando a identificação de tendências de longo prazo.
O parâmetro janela controla o tamanho da janela para a média móvel

```python
def analisar_tendencias_adocoes_medias_moveis(df_brasil, janela=3):
    """
    Cria gráficos de linha com médias móveis para suavizar as flutuações e identificar tendências de longo prazo.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.
        janela (int): Tamanho da janela para a média móvel.
    """

    dados_ano = df_brasil.groupby('NU_ANO_ADOCAO')['QT_ADOCAO'].sum().reset_index()
    dados_ano['MEDIA_MOVEL'] = dados_ano['QT_ADOCAO'].rolling(window=janela).mean()

    plt.figure(figsize=(10, 6))
    sns.lineplot(x='NU_ANO_ADOCAO', y='QT_ADOCAO', data=dados_ano, label='Adoções')
    sns.lineplot(x='NU_ANO_ADOCAO', y='MEDIA_MOVEL', data=dados_ano, label=f'Média Móvel ({janela} anos)')
    plt.title('Adoções ao Longo do Tempo com Média Móvel')
    plt.xlabel('Ano de Adoção')
    plt.ylabel('Quantidade de Adoções')
    plt.xticks(dados_ano['NU_ANO_ADOCAO'].unique()) # Garante que todos os anos sejam exibidos no eixo x
    plt.legend()
    plt.show()

# Supondo que seu DataFrame se chama df_brasil
analisar_tendencias_adocoes_medias_moveis(df_brasil)
```

```python
def analisar_tendencias_adocoes_medias_moveis_segmento(df_brasil, janela=3):
    """
    Cria gráficos de linha com médias móveis para suavizar as flutuações e identificar tendências de longo prazo por segmento.

    Args:
        df_brasil (pd.DataFrame): DataFrame com dados de adoção da Editora do Brasil.
        janela (int): Tamanho da janela para a média móvel.
    """

    segmentos = df_brasil['NM_SEGMENTO'].unique()

    for segmento in segmentos:
        dados_segmento_ano = df_brasil[df_brasil['NM_SEGMENTO'] == segmento].groupby('NU_ANO_ADOCAO')['QT_ADOCAO'].sum().reset_index()
        dados_segmento_ano['MEDIA_MOVEL'] = dados_segmento_ano['QT_ADOCAO'].rolling(window=janela).mean()

        plt.figure(figsize=(10, 6))
        sns.lineplot(x='NU_ANO_ADOCAO', y='QT_ADOCAO', data=dados_segmento_ano, label='Adoções')
        sns.lineplot(x='NU_ANO_ADOCAO', y='MEDIA_MOVEL', data=dados_segmento_ano, label=f'Média Móvel ({janela} anos)')
        plt.title(f'Adoções de "{segmento}" ao Longo do Tempo com Média Móvel')
        plt.xlabel('Ano de Adoção')
        plt.ylabel('Quantidade de Adoções')
        plt.xticks(dados_segmento_ano['NU_ANO_ADOCAO'].unique()) # Garante que todos os anos sejam exibidos no eixo x
        plt.legend()
        plt.show()

# Supondo que seu DataFrame se chama df_brasil
analisar_tendencias_adocoes_medias_moveis_segmento(df_brasil)
```
